```python
# Fully importing needed packages and modules

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns\

import sklearn
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from sklearn.model_selection import (
    train_test_split,
    cross_val_predict,
    GridSearchCV
)

from sklearn.preprocessing import (
    LabelEncoder,
    OneHotEncoder,
    OrdinalEncoder
)

from sklearn.metrics import(
    confusion_matrix,
    classification_report,
    accuracy_score
)

# Reading dataset
url="https://raw.githubusercontent.com/alishermutalov/praktikum-datasets/refs/heads/praktikum/mushrooms.csv"
df = pd.read_csv(url)

# javobni shu yerda yozing.
df = df.drop("veil-type", axis = 1) # veil-type – qoʻziqorin pardasi turi (barcha qiymatlar p, yaʼni partial (qisman) boʻlgani uchun foydasiz ustun).
df.head()

# Exploring dataset
print(df.info(), end = "\n\n")
print(df.shape, end = "\n\n")
df.describe()
```

# Corellation

```python
# Correletaion

encoder = LabelEncoder()

df_corr = df

for column in df.columns:
    df_corr[column] = encoder.fit_transform(df_corr[column])

corr_matrix = df_corr.corr().abs()
corr_matrix.style.background_gradient(cmap = "coolwarm")
```

```python
df_corr.corrwith(df_corr['class'], numeric_only=True).abs().sort_values(ascending=False)
```

# Dataset preparation for Machine Learning

```python
df['class'] = df['class'].replace({'p':1, 'e':0})

# Normalizing dataset

df_p = df[df['class'] == 1][:3916]
df_e = df[df['class'] == 0][:3916]
df_norm = pd.concat([df_p, df_e])

X = df.drop('class', axis = 1)
y = df['class']

# Train test split (70, 30)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=20)

# Pipline
pipeline = Pipeline([
    ('encoding', OneHotEncoder())
])

X_train_prepared = pipeline.fit_transform(X_train)

X_test_prepared = pipeline.transform(X_test)

X_full_prepared = pipeline.transform(X)

X_train_prepared.toarray()[0, :]
```

# 1st model. Logistic regression

```python
logistic_model = LogisticRegression(max_iter=100)

logistic_model.fit(X_train_prepared, y_train)

y_predict = logistic_model.predict(X_test_prepared)

# Accuracy score
print(f"Logistic regression accuraccy score : {accuracy_score(y_test, y_predict)}")

# Classification Report
print(f'Logistic regression report :\n{classification_report(y_test, y_predict)}\n')

print("Logistic regression confusion matrix:")
sns.heatmap(confusion_matrix(y_test, y_predict), annot=True)
plt.show()
```

# Cross Validation Predict. Logistic regression

```python
predict = cross_val_predict(estimator = logistic_model, X = X_full_prepared, y = y, cv = 5)

print(f'Cross validation report: \n{classification_report(y, predict)}')
```

# 2nd model. KNN

```python
# Finding the best K

knn = KNeighborsClassifier()
params = {'n_neighbors' : np.arange(1, 30)}
knn_gs = GridSearchCV(estimator = knn, param_grid=params, cv = 5)
knn_gs.fit(X_full_prepared, y)
```

```python
scores = knn_gs.cv_results_['rank_test_score']
```

```python
knn_gs.best_score_
```

```python
knn_gs.best_params_
```

```python
plt.figure(figsize = (12,5))
plt.plot(params['n_neighbors'], scores, color = 'red')
plt.title("The relation between K and error scoring")
plt.xticks(params['n_neighbors'])
plt.xlabel("K")
plt.ylabel("Error scores")
plt.grid()
plt.legend()
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors = 2)
knn.fit(X_train_prepared, y_train)
```

```python
y_predict = knn.predict(X_test_prepared)
```

```python
print(f'KNN accuracy score: {accuracy_score(y_test, y_predict)}')
print(f'KNN classification report : {classification_report(y_test, y_predict)}')
```

```python
print(f"KNN confusion matrix:\n")
sns.heatmap(confusion_matrix(y_test, y_predict), annot=True)
plt.show()
```

# Cross validation. KNN

```python
predict = cross_val_predict(estimator = knn, X = X_full_prepared, y = y, cv=5)

print(f'Crooss validation report:\n{classification_report(y, predict)}')
```
